```python
import pickle
import numpy as np
```

```python
pipe=pickle.load(open('pipe.pkl','rb'))
```

```python
pipe
```

```python
# Function to take user input and make predictions
def predict_survival():
    print("Enter passenger details for prediction:")

    pclass = int(input("Pclass (1, 2, or 3): "))
    age = float(input("Age: "))
    sibsp = int(input("Number of siblings/spouses aboard: "))
    parch = int(input("Number of parents/children aboard: "))
    fare = float(input("Fare: "))
    sex = input("Sex (male/female): ").strip().lower()
    embarked = input("Embarked (C/Q/S): ").strip().upper()

    # Create input array
    user_input = np.array([[pclass, sex, age, sibsp, parch, fare, embarked]])

    # Predict using the trained pipeline
    prediction = pipe.predict(user_input)

    # Output result
    if prediction[0] == 1:
        print("\nPrediction: The passenger **Survived** 🟢")
    else:
        print("\nPrediction: The passenger **Did Not Survive** 🔴")

# Run inference
predict_survival()
```
