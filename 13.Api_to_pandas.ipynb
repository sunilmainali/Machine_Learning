```python
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry

session = requests.Session()
retry = Retry(total=5, backoff_factor=2, status_forcelist=[500, 502, 503, 504])
session.mount('https://', HTTPAdapter(max_retries=retry))

df = pd.DataFrame()

for i in range(1, 503):
    try:
        response = session.get(
            f"https://api.themoviedb.org/3/movie/top_rated?api_key=88e7a3bb71da7b7c986849b8ed588e6a&language=en-US&page={i}",
            timeout=10
        )
        response.raise_for_status()
        temp_d = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity']]
        df = pd.concat([df, temp_d], ignore_index=True)
        time.sleep(7)  # Slow down requests
    except requests.exceptions.RequestException as e:
        print(f"Error on page {i}: {e}")
        time.sleep(7)
```
