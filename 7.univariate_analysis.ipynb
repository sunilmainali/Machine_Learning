```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('titan.csv')
```

```python
df.sample(5)
```

#### 1. Categorical Data

##### a.countplot

```python
sns.countplot(x=df['Embarked']).set_xticklabels(labels=['C', 'Q', 'S'],rotation=45)
```

sns.countplot(x=df['Survived']).set_xticklabels(labels=['0','1'],rotation=45)

#### b. PieChart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

#### 2. Numerical Data

##### a. Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'],bins=5)
```

##### b.distplot

```python
sns.distplot(df['Age'])
```

##### C.Boxplot

```python
sns.boxplot(df['Age'])
```
